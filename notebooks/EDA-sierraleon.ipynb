```python
#this is for Time Series Analysis task
import matplotlib.pyplot as plt

# Load dataset and parse Timestamp column
ds = pd.read_csv('../data/sierraleone-bumbuna.csv', parse_dates=['Timestamp'])

# Set Timestamp as index for easy time-series analysis
ds.set_index('Timestamp', inplace=True)

# Plot GHI, DNI, DHI, and Tamb
ds[['GHI', 'DNI', 'DHI', 'Tamb']].plot(figsize=(14, 7), alpha=0.8)
plt.title("Time Series of GHI, DNI, DHI, and Tamb")
plt.xlabel("Time")
plt.ylabel("Values")
plt.legend(loc='best')
plt.show()

# Add 'Month' and 'Hour' columns
ds['Month'] = ds.index.month
ds['Hour'] = ds.index.hour

# Plot average GHI, DNI, DHI by Month
ds.groupby('Month')[['GHI', 'DNI', 'DHI']].mean().plot(kind='bar', figsize=(12, 6))
plt.title("Average Solar Radiation by Month")
plt.ylabel("Radiation (W/m)")
plt.show()

# Plot average temperature (Tamb) by Hour
ds.groupby('Hour')['Tamb'].mean().plot(kind='line', figsize=(12, 6))
plt.title("Average Temperature by Hour of Day")
plt.ylabel("Temperature (C)")
plt.show()
```

```python
# Evaluate the impact of cleaning (using the 'Cleaning' column) on the sensor readings (ModA, ModB) over time.
import seaborn as sns


cleaned = ds[ds['Cleaning'] == 1]
not_cleaned = ds[ds['Cleaning'] == 0]

# Plot average ModA and ModB over time
plt.figure(figsize=(12, 6))
plt.plot(cleaned.index, cleaned['ModA'], label="ModA (Cleaned)", alpha=0.8)
plt.plot(not_cleaned.index, not_cleaned['ModA'], label="ModA (Not Cleaned)", alpha=0.8)
plt.title("Impact of Cleaning on ModA")
plt.xlabel("Time")
plt.ylabel("Sensor Reading (W/m)")
plt.legend()
plt.show()



# Compute correlation matrix
corr_matrix = ds[['GHI', 'DNI', 'DHI', 'TModA', 'TModB']].corr()

# Plot correlation matrix as a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix of Solar Radiation and Temperature")
plt.show()

#differentiate between Wind Conditions vs. Solar Irradiance
# Scatter plot: GHI vs. WS
plt.scatter(ds['WS'], ds['GHI'], alpha=0.6)
plt.title("Scatter Plot: Wind Speed (WS) vs. GHI")
plt.xlabel("Wind Speed (m/s)")
plt.ylabel("GHI (W/m)")
plt.show()

# Scatter plot: DNI vs. WSgust
plt.scatter(ds['WSgust'], ds['DNI'], alpha=0.6)
plt.title("Scatter Plot: Wind Gust Speed (WSgust) vs. DNI")
plt.xlabel("Wind Gust Speed (m/s)")
plt.ylabel("DNI (W/m)")
plt.show()
```

```python
from windrose import WindroseAxes
import matplotlib.pyplot as plt
import numpy as np

# Create a wind rose
plt.figure(figsize=(8, 8))
ax = WindroseAxes.from_ax()
ax.bar(ds['WD'], ds['WS'], normed=True, opening=0.8, edgecolor='white')
ax.set_legend(title="Wind Speed (m/s)")
plt.title("Wind Rose: Distribution of Wind Speed and Direction")
plt.show()



# Group by wind direction (rounded to nearest 10 degrees for simplicity)
ds['WD_rounded'] = ds['WD'].round(-1)
wind_speed_mean = ds.groupby('WD_rounded')['WS'].mean()

# Create a radial bar plot
angles = np.deg2rad(wind_speed_mean.index)  # Convert degrees to radians
values = wind_speed_mean.values

fig, ax = plt.subplots(figsize=(8, 8), subplot_kw={'projection': 'polar'})
bars = ax.bar(angles, values, color='skyblue', alpha=0.7)

ax.set_theta_zero_location("N")
ax.set_theta_direction(-1)
ax.set_title("Radial Bar Plot: Average Wind Speed by Direction")
plt.show()
```
